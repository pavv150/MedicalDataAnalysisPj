```python
#region  southwest = 0 southeast = 1 northwest = 2 northeast =3  문자 숫자로 변환
for i in range(len(df['region'])) :
    if df['region'][i] == 'southwest':
        df['region'][i] = 0
    elif df['region'][i] == 'southeast':
        df['region'][i] = 1
    elif df['region'][i] == 'northwast':
        df['region'][i] = 2
    else:
        df['region'][i] = 3
        
df
```

```python
# charges를 예측하기위해 변수대입
medical_input = df.drop('charges', axis=1)
medical_target = df['charges']

medical_target
```

```python
# train_test_split으로 데이터를 잘라줌
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    medical_input,medical_target,test_size=0.2 ,random_state=42)
```

```python
# 선형 회귀 모델
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

lr = LinearRegression().fit(x_train,y_train)
print(lr.score(x_train, y_train))
print(lr.score(x_test,y_test))
lr.predict(x_test)[:5]

y_pred_lr = lr.predict(x_test)
predict_values = lr.predict(x_train)

# 실제 값과 예측 값 사이의 차이 시각화
plt.scatter(y_test, y_pred_lr, alpha=0.5, label='Test Data')
plt.scatter(y_train, predict_values, alpha=0.5, label='Train Data')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Actual vs. Predicted Charges (LinearRegression)')
plt.legend()
plt.show()
```

```python
# 랜덤포레스트 모델
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)

print(rf.score(x_train, y_train))
print(rf.score(x_test,y_test))
rf.predict(x_test)[:5]
```

```python
# 랜덤포레스트 ,교차검증
from sklearn.model_selection import cross_validate

rf = RandomForestRegressor(n_jobs=-1, random_state=42,n_estimators=100)
scores = cross_validate(rf, x_train, y_train, return_train_score=True, n_jobs=-1)

np.mean(scores['train_score']), np.mean(scores['test_score'])
```

```python
# 그라디언트 부스팅 모델
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=100, random_state=43)
gb.fit(x_train,y_train)

print(gb.score(x_train, y_train))
print(gb.score(x_test,y_test))
gb.predict(x_test)[:5]
```
